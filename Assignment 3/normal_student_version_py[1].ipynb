```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import seaborn as sns
import pandas as pd

# Step 1 – Set mean and standard deviation
mu = 100       # Mean IQ
sigma = 15     # Standard deviation
sample_size = 10000

# Step 2 – Simulate normal distribution
iq_scores = np.random.normal(loc=mu, scale=sigma, size=sample_size)

# Step 3 – Compute actual sample mean and std deviation
mean_iq = np.mean(iq_scores)
std_iq = np.std(iq_scores)

# Step 4 – Prepare theoretical normal PDF
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
theoretical_pdf = norm.pdf(x, loc=mu, scale=sigma)

# Step 5 – Create a DataFrame
df = pd.DataFrame({'IQ': iq_scores})

# Step 6 – Plot simulated histogram and theoretical normal curve
plt.figure(figsize=(12, 6))
sns.histplot(df['IQ'], bins=50, stat="density", color='skyblue', label="Simulated Histogram", kde=False)
plt.plot(x, theoretical_pdf, 'r-', lw=2, label="Theoretical Normal PDF")
plt.axvline(mean_iq, color='green', linestyle='dashed', linewidth=1.5, label=f'Mean: {mean_iq:.2f}')
plt.axvline(mu, color='black', linestyle='dotted', linewidth=1.5, label=f'True Mean: {mu}')
plt.title("Normal Distribution Simulation of IQ Scores")
plt.xlabel("IQ Score")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

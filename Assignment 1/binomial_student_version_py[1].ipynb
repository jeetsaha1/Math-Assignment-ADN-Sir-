```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
import seaborn as sns
import pandas as pd

# TODO: Step 1 – Set the values for n, p, and number of simulations
n = 1000
p = 0.85
num_simulations = 1000

# TODO: Step 2 – Simulate Binomial distribution
simulated_successes = np.random.binomial(n=n, p=p, size=num_simulations)

# TODO: Step 3 – Compute mean and standard deviation
mean_simulated = np.mean(simulated_successes)
std_simulated = np.std(simulated_successes)

# TODO: Step 4 – Prepare theoretical binomial PMF
x = np.arange(min(simulated_successes), max(simulated_successes)+1)
theoretical_pmf = binom.pmf(x, n, p)

# TODO: Step 5 – Create a DataFrame for plotting
df = pd.DataFrame({'Simulated Successes': simulated_successes})

# TODO: Step 6 – Plot simulated results and theoretical PMF
plt.figure(figsize=(12, 6))
sns.histplot(df['Simulated Successes'], bins=30, stat="density", color='skyblue', label="Simulated Histogram")
plt.plot(x, theoretical_pmf, 'r-', lw=2, label="Theoretical Binomial PMF")
plt.axvline(mean_simulated, color='blue', linestyle='dashed', linewidth=1.5, label=f'Mean: {mean_simulated:.2f}')
plt.title("Vaccine Trial Simulation (Binomial)")
plt.xlabel("Number of Successful Outcomes")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
